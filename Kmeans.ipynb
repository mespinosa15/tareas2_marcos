```julia
using StatsPlots, DataFrames, Distributions, Clustering
```

```julia
data = vcat([0.3*randn(2) for i = 1:100],
    [[1,1] + 0.3*randn(2) for i = 1:100],
    [[1,-1] + 0.3*randn(2) for i = 1:100]);
xdata = zeros(100)
ydata = zeros(100)
for i in 1:100
    xdata[i] = data[i][1]
    ydata[i] = data[i][2]
end
```

```julia
scatter(xdata,ydata)
```

```julia
test = [xdata ydata];
```

```julia
test_matrix = permutedims(convert(Matrix{Float64}, test), [2, 1]);
```

```julia
result = kmeans(test_matrix, 3);
```

```julia
scatter(xdata, ydata, color = result.assignments)
```
